```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
iteration=10000
listX=[]
listY=[]
muX=3
muY=2
sigmaX=0.5
sigmaY=3
for i in range (iteration): 
    u=np.random.uniform(low=0,high=1)
    v=np.random.uniform(low=0,high=1)
    X=sigmaX*np.sqrt((-2)*np.log(u))*np.cos(2*np.pi*v) + muX
    listX.append(X)
    Y=sigmaY*np.sqrt((-2)*np.log(u))*np.sin(2*np.pi*v) + muY
    listY.append(Y)
```

```python
fig,ax=plt.subplots(figsize=(15, 4))
sb.distplot(listX)
sb.distplot(listY)
plt.title("Histogram of sampled realizations and approximated density using Box Muller Substitution")
fig,ax=plt.subplots(figsize=(15, 4))
sb.distplot(np.random.normal(muX,sigmaX,iteration),color="red")
sb.distplot(np.random.normal(muY,sigmaY,iteration),color="green")
ax.set_title("Histogram of sampled realizations and approximated density using built-in numpy gaussian function")
```

```python
fig,ax=plt.subplots(figsize=(15, 4))
plt.scatter(listX,listY)
fig,ax=plt.subplots(figsize=(15, 4))
mu=np.array([muX,muY])
sigma=np.array([[sigmaX,0],[0,sigmaY]])
dataX,dataY=np.random.multivariate_normal(mu,sigma,iteration).T
plt.scatter(dataX,dataY)
```
